```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt 
```

```python
dataset="/content/data setproject (1) (1).csv"
df=pd.read_csv(dataset)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Year_of_ study']= label_encoder.fit_transform(df['Year_of_ study'])
df['Year_of_ study'].unique()  
df['Living']= label_encoder.fit_transform(df['Living'])
  
df['Living'].unique()

df['Gender']= label_encoder.fit_transform(df['Gender'])
  
df['Gender'].unique()
df['Scholarship']= label_encoder.fit_transform(df['Scholarship'])
  
df['Scholarship'].unique()
df['Part_time_job']= label_encoder.fit_transform(df['Part_time_job'])
  
df['Part_time_job'].unique()
df['Mode_of_transportation']= label_encoder.fit_transform(df['Mode_of_transportation'])
  
df['Mode_of_transportation'].unique()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
x =df.iloc[:,:-1].values
y =df.iloc[:,-1].values
```

```python
#importing standard scalar from scikit-learn
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x= sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python

y_pred=model.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
print("Predicted value for training data:",model.score(x_train,y_train))
print("Training accuracy:",model.score(x_train,y_train)*100)
```

```python
print("Predicted value for testing data:",model.score(x_test,y_test))
print("Testing accuracy:",model.score(x_test,y_test)*100)
```

```python
accuracy=model.score(x,y)*100
print(accuracy)
```
