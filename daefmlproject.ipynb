```python
print(df.Sex.value_counts())
# Visualize frequency count of male / female for given titanic dataset using bar graph
df.Sex.value_counts().plot.bar()
```

```python
# Visualize frequency count of Age group of [21,20,19,18] using bar graph
print(df.loc[df.Age.isin([21,20,19,18]),['Age']].value_counts())
df.loc[df.Age.isin([21,20,19,18]),['Age']].value_counts().plot.bar()
```

```python
# Visualize frequency count of Age group of [21,20,19,18] using barh graph
df.loc[df.Age.isin([21,20,19,18]),['Age']].value_counts().plot.barh()
```

```python
#Display frequency count of Age group 30 to 50
print(df.loc[(df.Age<=21) & (df.Age>=18),['Age']].value_counts())
```

```python

print(df[df.Sex=='female'].loc[:,'Age'].max()) 
print(df[df.Sex=='male'].loc[:,'Age'].max()) 
print(df[df.Sex=='female'].loc[:,'Age'].mean()) 
print(df[df.Sex=='male'].loc[:,'Age'].mean()) 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.title('Gender Wise Ananlysis') # add title 
plt.xlabel("Gender of Passengers") # adds x -axis label 
plt.ylabel("Age of Passengers") # adds y -axis label
plt.bar(df.Sex,df.Age) # generates bar graph 
plt.show() 
```
